```python
import mne
import numpy as np
import os

# Folder containing your .mff files
data_folder = r"D:\Shivam MS\MI_final\MI_FINAL_filterred_mff"
all_files = [f for f in os.listdir(data_folder) if f.endswith(".mff")]

# Event mapping for motor imagery
imagination_events = {'ESLO': 7, 'ESRO': 9}

# EEG channels of interest
selected_channels = ['E36', 'E104','E30','E105']

# Epoch & filtering parameters
tmin, tmax = 0, 5           # extract 5 sec after event
sfreq = 500                 # sampling frequency (Hz)
window_size = 1             # 1 sec segments
step_size = 1               # non-overlapping
l_freq, h_freq = 8, 50      # band-pass range

X_list = []
y_list = []

for file in all_files:
    print(f"Processing file: {file}")
    raw_path = os.path.join(data_folder, file)
    
    # Load EEG data
    raw = mne.io.read_raw_egi(raw_path, preload=True)
    
    # Apply band-pass filter (in-memory)
    raw.filter(l_freq=l_freq, h_freq=h_freq, fir_design='firwin')
    
    # Extract events from annotations
    events, event_id = mne.events_from_annotations(raw)
    
    # Check if imagination events exist
    matching = [e for e in imagination_events.keys() if e in event_id]
    if not matching:
        print(f"No imagination events found in {file}, skipping...\n")
        continue
    
    # Create 5-sec epochs
    epochs = mne.Epochs(raw, events, event_id=imagination_events,
                        tmin=tmin, tmax=tmax, baseline=None, preload=True)
    epochs.pick_channels(selected_channels)
    
    # Get data and labels
    data = epochs.get_data()   # shape: (n_epochs, n_channels, n_times)
    labels = epochs.events[:, -1]
    
    n_samples_window = int(window_size * sfreq)
    n_samples_step = int(step_size * sfreq)
    
    # Segment each 5-sec epoch into 1-sec chunks
    for i_epoch, epoch_data in enumerate(data):
        start = 0
        while start + n_samples_window <= epoch_data.shape[1]:
            chunk = epoch_data[:, start:start + n_samples_window]
            X_list.append(chunk)
            y_list.append(labels[i_epoch])
            start += n_samples_step

# Combine all segments into arrays
X = np.stack(X_list)
y = np.array(y_list)

print("\n✅ Data segmentation complete.")
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Unique labels:", np.unique(y))
```

```python
# For LSTM: transpose to (samples, timesteps, channels)
X_lstm = np.transpose(X, (0, 2, 1))  # (n_samples, n_times, n_channels)

# Normalize per channel (mean=0, std=1)
X_norm = np.zeros_like(X_lstm, dtype=np.float32)
for ch in range(X_lstm.shape[2]):
    X_ch = X_lstm[:, :, ch]
    mean = X_ch.mean(axis=1, keepdims=True)  # mean per sample
    std = X_ch.std(axis=1, keepdims=True) + 1e-9  # avoid div by zero
    X_norm[:, :, ch] = (X_ch - mean) / std

X_lstm = X_norm
print("After normalization, shape:", X_lstm.shape)
```

```python
@tf.function
def real_cal(r, phi):
    return r * tf.math.cos(phi)

@tf.function
def imag_cal(r, phi):
    return r * tf.math.sin(phi)

@tf.function
def oscillator_loop(X_r, X_i, omegas, num_steps):
    # batch_size x timesteps X dim
    r_arr = tf.TensorArray(dtype=tf.float32, size=0, dynamic_size=True) # creates empty array to save r_t
    phi_arr = tf.TensorArray(dtype=tf.float32, size=0, dynamic_size=True)
    r_t = tf.ones((X_r.shape[0], X_r.shape[-1])) # Initializing r_t
    phis = tf.zeros((X_r.shape[0], X_r.shape[-1])) # Initislizing phi_t
    dt = 1/500
    input_scaler = 2
    beta=0.1

    for t in tf.range(num_steps):
        input_r = input_scaler*X_r[:,t,:]*tf.math.cos(phis)
        input_phi = input_scaler*X_i[:,t,:]*tf.math.sin(phis)
        r_t = r_t + ((1 - beta*tf.square(r_t)) * r_t + input_r) * dt
        phis = phis + (omegas - input_phi) * dt
        r_arr = r_arr.write(r_arr.size(), r_t)  #1000,1,2
        phi_arr = phi_arr.write(phi_arr.size(), phis)
    r_arr = tf.transpose(r_arr.stack(), [1, 0, 2])  # Changing dimensions to 1,1000,2
    phi_arr = tf.transpose(phi_arr.stack(), [1, 0, 2])
    return r_arr, phi_arr

class Hopf(tf.keras.layers.Layer):

    def __init__(self, units, num_steps, min_omega=8,
                 max_omega=50,train_omegas=False, **kwargs):
        super(Hopf, self).__init__(**kwargs)
        self.units = units
        self.num_steps = num_steps
        self.train_omegas=train_omegas
        self.max_omega=max_omega
        self.min_omega=min_omega
        # self.omegas = tf.linspace(min_omega, max_omega, self.units) * (2*3.1415)
        # self.omegas = tf.cast(tf.expand_dims(self.omegas, 0), 'float32')

        omega_init=tf.random.uniform((1,self.units),-1,1)
        self.omegas=tf.Variable(omega_init,trainable=self.train_omegas)

    def call(self, X_r, X_i):
        omega_intl = self.max_omega - self.min_omega
        omega_inp = tf.sigmoid(0.5*self.omegas) * omega_intl + self.min_omega


        r, phi = oscillator_loop(X_r, X_i, omega_inp , self.num_steps)
        z_real = real_cal(r, phi)
        z_imag = imag_cal(r, phi)
        return z_real, z_imag


duration = 500

class Model(tf.keras.Model):

    def __init__(self, units1, units2,units3,units4,units5, **kwargs):

        super(Model, self).__init__(**kwargs)

        self.d1_r = tf.keras.layers.Dense(units1,activation='relu') # 1st relu layer(real) with neuron=units1
        self.d1_i = tf.keras.layers.Dense(units1,activation='relu')# 1st relu layer(imag) with neuron=units1

        self.osc1 = Hopf(units2, num_steps=500,min_omega=8, max_omega=30,train_omegas=True) # 1st osc layer=units2=units1


        #self.d2_r = tf.keras.layers.Dense(units3,activation='relu') # 1st relu layer(real) with neuron=units1
        #self.d2_i = tf.keras.layers.Dense(units3,activation='relu')# 1st relu layer(imag) with neuron=units1

        #self.osc2 = Hopf(units4, num_steps=250,min_omega=7.9, max_omega=30.1,train_omegas=True) # 1st osc layer=units2=units1

        self.d_r = tf.keras.layers.Dense(units3,activation='tanh') # 1st relu layer(real) with neuron=units1
        self.d_i = tf.keras.layers.Dense(units3,activation='tanh')# 1st relu layer(imag) with neuron=units1
        
        
        self.dense = tf.keras.layers.Dense(units4, activation='relu') # 2nd relu layer,units3

        #self.d2_dr = tf.keras.layers.Dense(units4, activation='linear') #last tanh layer,units5
        #self.d2_i = tf.keras.layers.Dense(units4, activation='linear') #last tanh layer,units5
        # self.dense2= tf.keras.layers.Dense(units5, activation='relu') # 2nd relu layer,units3

        self.out_dense = tf.keras.layers.Dense(units5, activation='linear')# output node, units6, with tanh

    def call(self, X):

        out1_r = tf.keras.layers.TimeDistributed(self.d1_r)(X)
        out1_i = tf.keras.layers.TimeDistributed(self.d1_i)(X)

        z1_r, z1_i = self.osc1(out1_r, out1_i)




        #concat_inp=tf.concat([z1_r, z1_i],2)
        out2_r = tf.keras.layers.TimeDistributed(self.d_r)(z1_r)
        out2_i = tf.keras.layers.TimeDistributed(self.d_i)(z1_i)



        concat_inp=tf.concat([out2_r,out2_i],2)
        # lstmout=self.lstm(concat_inp)
        
        #out3_r = tf.keras.layers.TimeDistributed(self.d2_r)(out2_r)
        #out3_i = tf.keras.layers.TimeDistributed(self.d2_i)(out2_i)
        #out3=tf.concat([out3_r,out3_i],2)
        denseout=tf.keras.layers.TimeDistributed(self.dense)(concat_inp)
        # denseout2=tf.keras.layers.TimeDistributed(self.dense2)(denseout)
        out_final = tf.keras.layers.TimeDistributed(self.out_dense)(denseout)
        return out_final

model = Model(100,100,64,32,2)
optimizer = tf.keras.optimizers.Adam(0.01)


model.compile(optimizer, 'mse', metrics=['accuracy'])


```

```python
import numpy as np

def calculate_accuracy(pred_test, true_test):
    
    predicted_classes = np.argmax(pred_test, axis=1)  
    
    
    true_classes = np.argmax(true_test, axis=1) 
    
    
    accuracy = np.mean(predicted_classes == true_classes) 
    return accuracy


accuracy = calculate_accuracy(pred_test, true_test)
print(f"Accuracy: {accuracy:.4f}")
```
